```python
#importing modules
import numpy as numpy
import yfinance as yf
```

```python
#looking at your data/getting dataset
data=yf.download('AAPL', start='2024-06-10', end="2025-06-11")
print(data.tail())
```

```python
#selecting independent and dependent variables
data['Return'] = data['Close'].pct_change() # calcs changing price from one day to the next --> creates a new column
data['MA10'] = data['Close'].rolling(window=10).mean() #takes the mean percent change over a set of 5 days --> creates a new column for moving average
data['Target'] = data['Close'].shift(-1) > data['Close'] #target variable is dependent variable
#shift(-1) is tomorrow's closing price
#if tomorrow's is greater than today's then return True and vice versa
#makes this new column just true or false
data['Target'] = data['Target'].astype(int)  # 1 = Up, 0 = Down
```

```python
#split into training and testing data
from sklearn.model_selection import train_test_split

X = data[['Return', 'MA10']] #x is always the features
y = data['Target'] #dependent variables --> 1 and 0

# Drop NaN values before splitting
X = X.dropna()
y = y.loc[X.index] # Align y with the cleaned X

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.26, random_state=42)
```

```python
#Visualize Data
print(X)
print(y)
X.isna().sum()
X=X.dropna()
y=y.dropna()
X.isna().sum()
y.isna().sum()
print(X_train)
print(X_test)
print(y_train)
print(y_test)
X_train.isna().sum()
X_test.isna().sum()
y_train.isna().sum()
y_test.isna().sum()
```

```python
#make the models
from sklearn.linear_model import LogisticRegression #binary classification
from sklearn.ensemble import RandomForestClassifier #binary classification
logModel = LogisticRegression()
logModel.fit(X_train, y_train)
classifier=RandomForestClassifier(n_estimators=100,random_state=26,oob_score=True) #oob means using the random trees that have the algorithm but were not tested with the training data, so prevents overfitting.
classifier.fit(X_train, y_train)
```

```python
#look at model accuracy
logPred=logModel.predict(X_test)
from sklearn.metrics import accuracy_score
print("Accuracy Score of Logistic Regression Model:",accuracy_score(y_test, logPred))

forestPred=classifier.predict(X_test)
print("Accuracy Score of Random Forests:", accuracy_score(y_test, forestPred))
print("Accuracy of OOB Score for Random Forests:",classifier.oob_score_)
```
